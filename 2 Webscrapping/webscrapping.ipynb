```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpages = []
soup_lis = []
for i in range(1, 334):
    webpages.append(requests.get(f'https://www.ambitionbox.com/list-of-companies?page={i}', headers=headers).text)
    soup_lis.append(BeautifulSoup(webpages[i-1], 'lxml'))
```

```python
name = []
rating = []
reviews = []
ctype = []
hq = []
old = []
employees = []
cinfo = [ctype, hq, old, employees]
for item in soup_lis:
    company = item.find_all('div', class_='company-content-wrapper')
    for i in company:
        k = 0
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p', class_='rating').text.strip())
        reviews.append(i.find('a', class_='review-count').text.strip())
        for j in cinfo:
            try:
                j.append(i.find_all('p', class_='infoEntity')[k].text.strip())
            except IndexError:
                j.append('')
            k = k+1
    d = {'name':name, 'rating':rating, 'reviews':reviews, 'ctype':ctype, 'hq':hq, 'old':old, 'employees':employees}
datasets = pd.DataFrame(d)
datasets
```

```python
datasets.to_csv('AmbitionBox_Datasets.csv')
```
